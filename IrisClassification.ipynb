## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

## Load the Dataset

```python
df = pd.read_csv("Iris.csv")
df
```

```python
# Printing the summary of a DataFrame

df.info()
```

```python
#Displaying the number of samples for each species in Dataset

df["Species"].value_counts()
```

## Data Preprocessing

```python
# Dropping Id column as it is not significant here

df=df.drop(columns=["Id"])
df.head(10)
```

```python
df.columns
```

```python
# Checking for null values

df.isnull().sum() 
```

```python
# Converting class labels into numerical form

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Species']=le.fit_transform(df['Species'])
df["Species"]
```

## Exploratory Data Analysis (EDA)

```python
# Creating dictionary to map hue values

hue_mapping = {0: 'Setosa', 1: 'Versicolor', 2: 'Virginica'}

# Mapping the "Species" column to the desired names

df['Species'] = df['Species'].map(hue_mapping)

# Creating pairplot with hue mapping and different diag_kind

sns.pairplot(df, hue='Species', diag_kind='kde')
plt.show()
```

## Correlation Matrix

```python
# Converting the "Species" column to numeric and then perform the correlation, we use the 'LabelEncoder'
from sklearn.preprocessing import LabelEncoder

# Creating LabelEncoder
le = LabelEncoder()

# Fitting and transforming the "Species" column to convert species names to integers
df['Species'] = le.fit_transform(df['Species'])

# Computing the correlation matrix
correlation_matrix = df.corr()

# Displaying the correlation matrix
print(correlation_matrix)
```

```python
# Displaying the correlation matrix using a heatmap

corr = correlation_matrix
fig, ax = plt.subplots(figsize=(5, 4))
sns.heatmap(corr, annot=True, ax=ax, cmap='viridis')
```

## Model Training

```python
# Spliting the dataset into training and testing sets

from sklearn.model_selection import train_test_split
X = df.drop(columns=['Species'])
Y = df['Species']
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.40)
```

```python
# Initializing logistic regression model

from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
# Model fitting

model.fit(x_train,y_train)

# Model accuracy

print("Accuracy(Logistic Regression): ",model.score(x_test,y_test)*100)
```
